# Compare CO2 and cation-based estimates of CDR
This script compares CDR calculated via SCEPTER CO2 fluxes with potential CDR calculated via SCEPTER cation fluxes. We use a calcite feedstock example to restrict our focus to $\text{Ca}^{2+}$.

```python
# ---------------------------------------------------
# 
# Compute the CDR fluxes based on the aq-pco2 outputs
# and the aq-cation outputs for comparison
# 
# ---------------------------------------------------
import numpy as np
import pandas as pd
from matplotlib import gridspec
import matplotlib.pyplot as plt
import pickle
import os
import re
import fsspec
import s3fs
```

## Helper function
Read in and convert .txt files to pandas dataframes

```python
def preprocess_txt(
    outdir: str,
    runname: str,
    fn: str,
    run_subdir: str = "flx",
    map_numeric: bool = True,
    S3on: bool=False,
) -> pd.DataFrame:
    """
    Convert SCEPTER output .txt files to pandas DataFrame for further
    analysis

    Parameters
    ----------
    outdir : str
        path to the SCEPTER output directory. Usually something like 'my/path/SCEPTER/scepter_output'
    runname : str
        name of the SCEPTER run (equivalent to the directory within outdir). Generally <RUNNAME>_field.
    fn : str
        name of the file in the flux directory to input
    run_subdir : str
        name of the subdirectory in the run output directory to find the file (either "prof" or "flux")
        default is "flx". 
    map_numeric : bool
        if True, all columns are mapped to numeric values. This should only be false for dust.txt which has 
        columns for the dust species

    Returns
    -------
    pd.DataFrame
        same format as the SCEPTER .txt file
    """
                    
    data = []  # Initialize a list to store the processed data

    # Initialize a flag to determine if we are reading the header
    is_header = True

    # get file path
    file_path = os.path.join(outdir, runname, run_subdir, fn)

    # Read the file line by line and process the data
    if S3on:
        with fs.open(file_path, 'r') as file:
            for line in file:
                line = line.strip()  # Remove leading/trailing whitespace
                if is_header:
                    # Split the first line into column names
                    column_names = re.split(r'\s+', line)
                    is_header = False
                else:
                    # Split the other lines into data values
                    values = re.split(r'\s+', line)
                    data.append(values)
    else:
        with open(file_path, 'r') as file:
            for line in file:
                line = line.strip()  # Remove leading/trailing whitespace
                if is_header:
                    # Split the first line into column names
                    column_names = re.split(r'\s+', line)
                    is_header = False
                else:
                    # Split the other lines into data values
                    values = re.split(r'\s+', line)
                    data.append(values)

    # Create a DataFrame with the processed data and set column names
    df = pd.DataFrame(data, columns=column_names)
    # apply pd.to_numeric to all columns using the "map" method
    if map_numeric:
        df = df.map(pd.to_numeric)
    # return
    return df
```

## Read in data 
These are 15-year runs with annual-mean climate forcing and no fertilizer. We're comparing a calcite application `case` (100 ton/ha/yr) to a `ctrl` scenario with no dust application. 

For each run (`case` and `ctrl`) we read in two files:
1. `flx_gas-pco2.txt`: for computing CDR based on carbon fluxes
2. `flx_aq-ca.txt`: for computing CDR based on cations (we're using a calcite for simplicity so we can just focus on calcium fluxes)

```python
# --- read in data
# set fertilization level ("no" or "hi")
fertlevel = "low"   # ["no" | "hi"]
# set directory
# outdir = "/home/tykukla/SCEPTER/scepter_output"
outdir = "s3://carbonplan-carbon-removal/SCEPTER/scepter_output_scratch/"
# read in 
# runname_ctrl = f"{fertlevel}Fert_cc__site_311a_app_0p0_psize_1_cc_field_tau15p0"
# runname_case = f"{fertlevel}Fert_cc__site_311a_app_10000p0_psize_150_cc_field_tau15p0"
runname_ctrl = f"{fertlevel}Fert_cc_basev9_site_311a_app_0p0_psize_1_cc_field_tau15p0"
runname_case = f"{fertlevel}Fert_cc_basev9_site_311a_app_10000p0_psize_150_cc_field_tau15p0"

file_subdir = "flx"  # subdir within case dir where the flux files exist
fn_pco2 = "int_flx_gas-pco2.txt"
fn_aqca = "int_flx_aq-ca.txt"

# --- check for s3
if outdir.startswith("s3://"): # then bring it in from s3
    S3on = True
    import s3fs
    fs = s3fs.S3FileSystem()
    s3_path = fs.exists(outdir)
else:
    S3on = False
# ********************************************************************************
# --- [1] CONTROL RUN
dfctrl_aqca = preprocess_txt(outdir, runname_ctrl, fn_aqca, file_subdir, True, S3on)
dfctrl_pco2 = preprocess_txt(outdir, runname_ctrl, fn_pco2, file_subdir, True, S3on)
# --- [2] CASE RUN
dfcase_aqca = preprocess_txt(outdir, runname_case, fn_aqca, file_subdir, True, S3on)
dfcase_pco2 = preprocess_txt(outdir, runname_case, fn_pco2, file_subdir, True, S3on)
# ********************************************************************************
```

## Convert pCO2 data to time-integrated CO2 fluxes
The pCO2 data are in $(\text{mol}\ \text{m}^{2}\ \text{yr}^{-1})$ and we want to convert to $(\text{ton}\ \text{ha}^{-1})$. We do this in two steps (an implied step is that 1 mole of bicarbonate equals 1 mole of CO2):

1. Convert $(\text{mol}\ \text{m}^{2}\ \text{yr}^{-1})$ to $(\text{ton}\ \text{ha}^{-1}\ \text{yr}^{-1})$
$$\text{conversion factor} = \frac{44.01\ \text{grams CO2}}{1\ \text{mole}} \frac{1\ \text{ton}}{1\text{E}6\ \text{gram}} \frac{1\text{E}4\ \text{m}^{2}}{1\ \text{ha}}$$  

2. Multiply by time to convert $(\text{ton}\ \text{ha}^{-1}\ \text{yr}^{-1})$ to $(\text{ton}\ \text{ha}^{-1})$

```python
# *****************************************
# define unit conversion constants
# for mol/m2/yr to ton/ha/yr
g_mol = 44.01
ton_g = 1 / 1e6   # [ton g-1]
m2_ha = 10e3      # [m2 ha-1]
conv_factor = g_mol * ton_g * m2_ha 
# *****************************************

# apply the conversion to all columns except 'time'
# --- [1] CONTROL RUN
dfctrl_pco2 = dfctrl_pco2.apply(lambda x: x * conv_factor * dfctrl_pco2['time'] if x.name not in ['time'] else x).copy()
# --- [2] CASE RUN
dfcase_pco2 = dfcase_pco2.apply(lambda x: x * conv_factor * dfcase_pco2['time'] if x.name not in ['time'] else x).copy()
```

## Convert cation data to time-integrated CO2potential fluxes
The cation flux data are in $(\text{mol}\ \text{m}^{2}\ \text{yr}^{-1})$ and we want to convert to $(\text{ton}\ \text{ha}^{-1})$. We do this in two steps that are almost identical to the pCO2 conversion above. The key difference is we multiply by two to convert cation flux to DIC flux. 

1. Convert $(\text{mol}\ \text{m}^{2}\ \text{yr}^{-1})$ to $(\text{ton}\ \text{ha}^{-1}\ \text{yr}^{-1})$
$$\text{conversion factor} = \frac{2\ \text{mole HCO3}^{-}}{1\ \text{mole Ca}^{2+}} \frac{44.01\ \text{grams CO2}}{1\ \text{mole}} \frac{1\ \text{ton}}{1\text{E}6\ \text{gram}} \frac{1\text{E}4\ \text{m}^{2}}{1\ \text{ha}}$$  

2. Multiply by time to convert $(\text{ton}\ \text{ha}^{-1}\ \text{yr}^{-1})$ to $(\text{ton}\ \text{ha}^{-1})$


```python
# ***********************************************
# define unit conversion constants
# for mol/m2/yr to ton/ha/yr
ca_charge = 2
g_mol = 44.01
ton_g = 1 / 1e6   # [ton g-1]
m2_ha = 10e3      # [m2 ha-1]
conv_factor = ca_charge * g_mol * ton_g * m2_ha 
# ***********************************************

# apply the conversion to all columns except 'time'
# --- [1] CONTROL RUN
dfctrl_aqca = dfctrl_aqca.apply(lambda x: x * conv_factor * dfcase_pco2['time'] if x.name not in ['time'] else x).copy()
# --- [2] CASE RUN
dfcase_aqca = dfcase_aqca.apply(lambda x: x * conv_factor * dfcase_pco2['time'] if x.name not in ['time'] else x).copy()
```

## Compute pco2 CDR
All CDR fluxes use the terms of the soil column mass balance equation:
$$0 = -\text{adv} - \text{dif} - [\text{sources}] - \text{tflx}$$

where `adv` is the flux out of the base of the column, `dif` is the flux out of the top, `sources` are fluxes into the column, and `tflx` is the change in column storage. Positive values indicate fluxes out of system, negative is into the system. 

In our simulations, we just have one inorganic source (`cc`) and one organic source (`g2`) so we can rewrite this to be:
$$0 = -\text{adv} - \text{dif} - \text{cc} - \text{g2} - \text{tflx}$$

We calculate CDR by comparing the `case` to the `ctrl` simulations. For example, we can compute the decrease in emissions at the top of the column (diffusive CDR) or the increase in DIC that fluxes out the base of the column (advective CDR). If there is no change in storage ($\text{tflx}=0$), these two fluxes should be identical (I think?)

### Diffusive CDR
This is the change in diffusion out the top of the column that is caused by mineral application. Since `dif` is positive out of the column (representing emissions) we can start with the decrease in emissions represented as:
$$-\text{dif}_\text{case} + \text{dif}_\text{ctrl}$$

Reorganizing the above equations gives us:
$$-\text{dif}_\text{case} + \text{dif}_\text{ctrl} = \Delta\text{adv} + \Delta\text{cc} + \Delta\text{g2} + \Delta\text{tflx}$$

where every $\Delta\text{X}$ term equals $\text{X}_\text{case}-\text{X}_\text{ctrl}$.

To get the contribution owed to mineral application we need to omit the change in diffusion due to respiration (`g2`), giving us:
$$\text{CDR}_\text{dif} = -\text{dif}_\text{case} + \text{dif}_\text{ctrl} - \Delta\text{g2}$$

We can re-write this as the sum of two components, a CDR contribution owed to the change in diffusion ($-\text{dif}_\text{case} + \text{dif}_\text{ctrl}$), and one owed to the change in respiration ($-\Delta\text{g2}$):
$$\text{CDR}_\text{dif} = \text{dif component} + \text{g2 component}$$

This framing reveals that the respiration flux can increase CDR if the `g2 component` is $>0$. To ensure we don't count those avoided emissions, we write the final equation:
$$\text{CDR}_\text{dif} = \text{dif component} + \text{MIN}(\text{g2 component}, 0)$$

(Note that the `cc` term is the carbon flux owed to carbonate mineral dissolution (e.g., the "fossil" carbon flux). I'm not sure whether we should account for that in our $\text{CDR}_\text{dif}$ calculation. My reason for omitting it here is that the carbon released from `cc` is generally balanced by $\text{Ca}^{2+}$ and is more likely to be advected out. But I don't know how sound this reasoning is / where the assumptions break!)

### Advective CDR
To calculate the advective flux, we do the same thing but instead of accounting for the change in the respiration contribution, we omit the `cc` contribution. Here, positive values represent carbon moving away from the atmosphere (unlike `dif`), so we can write the change in column advection such that positive values are CDR as:

$$\text{adv}_\text{case} - \text{adv}_\text{ctrl} = -\Delta\text{dif} -\Delta\text{g2} -\Delta\text{cc} -\Delta\text{tflx}$$

We omit `cc` by adding the $\Delta\text{cc}$ term to both sides, which gives us $\text{CDR}_\text{adv}$

$$\text{CDR}_\text{adv} = \text{adv}_\text{case} - \text{adv}_\text{ctrl} + \Delta\text{cc}$$

However, this equation counts a decrease in cc dissolution (or increase in cc precipitation) as advective CDR. We can restrict the advective term to just the advective flux without a positive CDR contribution from cc dissolution by:

$$\text{CDR}_\text{adv, noSIC} = \text{adv}_\text{case} - \text{adv}_\text{ctrl} + \text{MIN}(\Delta\text{cc}, 0)$$


```python
# --- pco2 CDR fluxes
# dataframe with the CDR components
dfcdr_pco2 = pd.DataFrame({
                'time': dfcase_pco2['time'],
                'dif_component': -dfcase_pco2['dif']+dfctrl_pco2['dif'],
                'g2_component': -(dfcase_pco2['g2']-dfctrl_pco2['g2']),
                'adv_component': dfcase_pco2['adv']-dfctrl_pco2['adv'],
                'D_g2': dfcase_pco2['g2']-dfctrl_pco2['g2'],
                'D_cc': dfcase_pco2['cc']-dfctrl_pco2['cc'],
                'D_tflx': dfcase_pco2['tflx']-dfctrl_pco2['tflx'],
                })

# compute diffusive CDR
dfcdr_pco2['cdr_dif'] = dfcdr_pco2['dif_component'] + np.minimum(dfcdr_pco2['g2_component'], 0)
dfcdr_pco2['cdr_adv'] = dfcdr_pco2['adv_component'] + np.minimum(dfcdr_pco2['D_cc'], 0)
dfcdr_pco2
```

```python
# --- compare the different components (case vs ctrl)
fig = plt.figure(figsize=(10,8))
gs = gridspec.GridSpec(2, 2, figure=fig, hspace=0.3) # 2x2 

# subplot 1: diffusive fluxes
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(dfcase_pco2['time'], dfcase_pco2['dif'], label='case')
ax1.plot(dfctrl_pco2['time'], dfctrl_pco2['dif'], label='ctrl')
plt.xlabel('year')
plt.ylabel('integrated flux (ton ha-1)')
plt.legend()
plt.title("diffusive flux");

# subplot 2: respiration fluxes
ax2 = fig.add_subplot(gs[0,1])
ax2.plot(dfcase_pco2['time'], dfcase_pco2['g2'], label='case')
ax2.plot(dfctrl_pco2['time'], dfctrl_pco2['g2'], label='ctrl', linestyle='dashed')
plt.xlabel('year')
plt.ylabel('integrated flux (ton ha-1)')
plt.legend()
plt.title("g2 flux");

# subplot 3: adv fluxes
ax3 = fig.add_subplot(gs[1,0])
ax3.plot(dfcase_pco2['time'], dfcase_pco2['adv'], label='case')
ax3.plot(dfctrl_pco2['time'], dfctrl_pco2['adv'], label='ctrl')
plt.xlabel('year')
plt.ylabel('integrated flux (ton ha-1)')
plt.legend()
plt.title("advective flux");

# subplot 4: cc fluxes
ax4 = fig.add_subplot(gs[1,1])
ax4.plot(dfcase_pco2['time'], dfcase_pco2['cc'], label='case')
ax4.plot(dfctrl_pco2['time'], dfctrl_pco2['cc'], label='ctrl')
plt.xlabel('year')
plt.ylabel('integrated flux (ton ha-1)')
plt.legend()
plt.title("cc flux");
```

```python
# --- compare the different components (dif vs adv cdr)
fig = plt.figure(figsize=(12,4))
gs = gridspec.GridSpec(1, 3, figure=fig, hspace=0.3) # 2x2 

# subplot 1: components of dif_cdr
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(dfcdr_pco2['time'], dfcdr_pco2['dif_component'], label='dif_component')
ax1.plot(dfcdr_pco2['time'], dfcdr_pco2['g2_component'], label='g2_component')
plt.xlabel('year')
plt.ylabel('integrated flux (ton ha-1)')
plt.legend()
plt.title("cdr_dif components");

# subplot 2: components of adv_cdr
ax3 = fig.add_subplot(gs[0,1])
ax3.plot(dfcdr_pco2['time'], dfcdr_pco2['adv_component'], label='adv_component')
ax3.plot(dfcdr_pco2['time'], dfcdr_pco2['D_cc'], label='cc_component')
plt.xlabel('year')
plt.ylabel('integrated flux (ton ha-1)')
plt.legend()
plt.title("cdr_adv components");

# subplot 3: dif vs adv cdr
ax4 = fig.add_subplot(gs[0,2])
ax4.plot(dfcdr_pco2['time'], dfcdr_pco2['cdr_dif'], label='cdr_dif')
ax4.plot(dfcdr_pco2['time'], dfcdr_pco2['cdr_adv'], label='cdr_adv')
plt.title("CDR calculated with pCO2 flxs")
plt.xlabel('year')
plt.ylabel('integrated CDR (ton ha-1)')
plt.legend();
```

## Compute cation-derived CDR 
We're going to compute two kinds of cation-derived CDR: $\text{CDR}_\text{adv}$ is the carbon removal potential due to cations flushed out the base of the soil column, and $\text{CDR}_\text{tot}$ is the carbon removal potential due to cations flushed out the base plus those added to the soil column.

The mass balance equation is fundamentally the same as the pco2 files, but the sources are different. Here, we can write the mass balance equation as: 

$$0 = -\text{adv} - \text{dif} - [\text{carbonate sources}] - [\text{other mineral sources}] - \text{tflx}$$

We split the minerals into carbonate and other because the CO2 potential of carbonate minerals per cation*charge released is half that of other mineral sources. The `dif` flux is negligible because the cations are not diffusing upward out of the soil column. In our simulation, cc is the only carbonate source and other mineral contributions are negligible. Given this, we can simplify to: 

$$0 = -\text{adv} - \text{cc} - \text{tflx}$$

Re-arranging, the advected flux is balanced by `cc` and `tflx`. And the carbon removal potential of the advected flux accounts for the stoichiometry of carbonate minerals, giving us 

$$\text{removal potential} = -\text{cc}/{2} - \text{tflx}$$

### Advective CDR
We get potential CDR by looking at the change in the removal potential from the `case` to the `control`, giving us:

$$\text{CDR}_\text{adv} = -\Delta\text{cc}/2 - \Delta\text{tflx}$$

where, as before, $\Delta\text{X}=\text{X}_\text{case} - \text{X}_\text{ctrl}$. Unlike the pco2 calculation, I'm not using any $MIN()$ logic here. If more cations are stored in the soil due to ERW, then $\Delta\text{tflx}$ will be positive and $\text{CDR}_\text{adv}$ is reduced. We could use $MAX()$ logic so the $\Delta\text{tflx}$ term is strictly positive. I think by not doing that we're essentially assuming any loss in soil cation storage means the release of cations sourced from our feedstock. (It wouldn't matter in this case since $\Delta\text{tflx}$ is positive anyway). 


### Total CDR
To get total CDR potential, we omit the storage flux. This means it doesn't matter what happens to storage in the soil column, CDR will only respond to cations released from `cc`. 

$$\text{CDR}_\text{tot} = -\Delta\text{cc}/2$$


```python
# --- pco2 CDR fluxes
# dataframe with the CDR components
dfcdr_aqca = pd.DataFrame({
                'time': dfcase_aqca['time'],
                'cc_component': -(dfcase_aqca['cc']-dfctrl_aqca['cc'])/2,
                'tflx_component': -(dfcase_aqca['tflx']-dfctrl_aqca['tflx']),
                'D_cc': dfcase_aqca['cc']-dfctrl_aqca['cc'],
                'D_tflx': dfcase_aqca['tflx']-dfctrl_aqca['tflx'],
                })

# compute diffusive CDR
dfcdr_aqca['cdr_adv'] = dfcdr_aqca['cc_component'] + dfcdr_aqca['tflx_component']
dfcdr_aqca['cdr_tot'] = dfcdr_aqca['cc_component'] 
dfcdr_aqca
```

```python
# --- compare the different components (case vs ctrl)
fig = plt.figure(figsize=(12,4))
gs = gridspec.GridSpec(1, 3, figure=fig, wspace=0.3) # 2x2 

# subplot 1: advective fluxes
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(dfcase_aqca['time'], dfcase_aqca['adv'], label='case')
ax1.plot(dfctrl_aqca['time'], dfctrl_aqca['adv'], label='ctrl')
plt.xlabel('year')
plt.ylabel('integrated flux (ton ha-1)')
plt.legend()
plt.title("advective ca2+ flux");

# subplot 2: cc fluxes
ax2 = fig.add_subplot(gs[0,1])
ax2.plot(dfcase_aqca['time'], dfcase_aqca['cc'], label='case')
ax2.plot(dfctrl_aqca['time'], dfctrl_aqca['cc'], label='ctrl', linestyle='dashed')
plt.xlabel('year')
plt.ylabel('integrated flux (ton ha-1)')
plt.legend()
plt.title("cc ca2+ flux");

# subplot 3: tflx fluxes
ax3 = fig.add_subplot(gs[0,2])
ax3.plot(dfcase_aqca['time'], dfcase_aqca['tflx'], label='case')
ax3.plot(dfctrl_aqca['time'], dfctrl_aqca['tflx'], label='ctrl')
plt.xlabel('year')
plt.ylabel('integrated flux (ton ha-1)')
plt.legend()
plt.title("delta storage ca2+ flux");
```

```python
# --- compare the different components (adv vs tot cation cdr)
fig = plt.figure(figsize=(10,4))
gs = gridspec.GridSpec(1, 2, figure=fig, hspace=0.3) # 2x2 

# subplot 1: components of adv_cdr
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(dfcdr_aqca['time'], dfcdr_aqca['cc_component'], label='cc_component')
ax1.plot(dfcdr_aqca['time'], dfcdr_aqca['tflx_component'], label='tflx_component', linestyle='dashed')
plt.xlabel('year')
plt.ylabel('integrated flux (ton ha-1)')
plt.legend()
plt.title("cdr components");

# subplot 2: adv vs tot cdr
ax3 = fig.add_subplot(gs[0,1])
ax3.plot(dfcdr_aqca['time'], dfcdr_aqca['cdr_tot'], label='cdr_tot')
ax3.plot(dfcdr_aqca['time'], dfcdr_aqca['cdr_adv'], label='cdr_adv')
plt.xlabel('year')
plt.ylabel('integrated flux (ton ha-1)')
plt.legend()
plt.title("adv vs tot cdr");
```

```python
# ------------------------------------------------------
```
